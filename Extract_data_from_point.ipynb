```python
import os
import glob
import rasterio
from pyproj import Transformer
import numpy as np
import pandas as pd
```

```python
lon = 14.333
lat = 41.033
```

```python
# List of GeoTIFF files to extract data from
files = glob.glob("*_ET0.tif")

# print (files)
# Create an empty list to store the data
data = []

# Loop through each file
for file in files:
    with rasterio.open(file) as rds:
        transformer = Transformer.from_crs("EPSG:4326", rds.crs, always_xy=True)
        xx, yy = transformer.transform(lon, lat)
        
        # Extract data at the point
        value = list(rds.sample([(xx, yy)]))[0]
#         value = src.sample([(x, y)])[0][0]
        # Add the value and file name to the data list
        data.append({'file': file, 'value': value})

# Convert the data to a pandas dataframe
df = pd.DataFrame(data)

# Save the dataframe to a CSV file
df.to_csv('ET0_atpoint.csv', index=False)
df.to_excel("ET0_atpoint.xlsx", index=False)
```
